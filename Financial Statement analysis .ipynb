```python
### USe of Numpy to Analyze Financial Statement  # reference : www.superdatascience.com

#Data 
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]

#profit = revenue-expenses in 1K dollars


import numpy as np
revenue_a = np.array(revenue)
expenses_a = np.array(expenses)
profit = revenue_a - expenses_a
print(profit)

#profit_after_tax = (profit - profit*(30//100)) in 1K dollars

profit_atax = (profit - profit * 0.3)

for i in range(len(profit)):
    if (profit[i]<0):
        profit_atax[i] = profit[i]
print(profit_atax)

#profit_margin = profit_after_tax/revenue in percentage no decimals
profit_margin = profit_atax / revenue_a *100
print(np.around(profit_margin))


#good_month = profit_after_tax>avg(profit_after_tax) in T/F
good_month = profit_atax > np.mean(profit_atax)
print(good_month)

#bad_months profit_after_tax<avg T/F
bad_month = profit_atax < np.mean(profit_atax)

print(bad_month)


#best_month max profit_after_tax T/F
best_month = profit_atax == np.max(profit_atax)
print(best_month)

#worst_month min profit_after_tax T/F
worst_month = profit_atax == np.min(profit_atax)
print(worst_month)



```

```python
#Data 
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]

#1: Calculating Profit (revenue-expenses)
profit = []
for i in range (0,len(revenue)):
    profit.append(revenue[i] - expenses[i])
print("profit:", profit)

#2: Calculating tax (profit *30%)
tax = [round(i * 0.3, 2)for i in profit]
print("tax:", tax)

#3: Calculating profit after tax (profit after tax= Profit-tax)
profit_atax = []
for i in range (0,len(profit)):
    profit_atax.append(profit[i] - tax[i])
print("profit_atax:", profit_atax)

#4: Calculating profit margine (profit after tax/revenue) in percent
profit_margin = []
for i in range (0,len(profit_atax)):
    profit_margin.append( profit_atax[i]/revenue[i])
profit_margin = [round(i*100,2) for i in profit_margin]
print("profit_margin:", profit_margin)

#5: Calculating mean of profit after tax

profit_atax
mean = sum(profit_atax) / len(profit_atax)
print("Mean:", mean)

#6:Good months (profit a_tax > mean)
Good_months = []
for i in range (0, len(profit_atax)):
    Good_months.append(profit_atax[i] > mean)
print("Good months:", Good_months)

#7: Bad months (profit a_tax < mean)
Bad_months = []
for i in range (0, len(profit_atax)):
    Bad_months.append(profit_atax[i] < mean)
print("Bad months:", Bad_months)


#8: Best Month 
best_month = []
for i in range (0, len(profit_atax)):
    best_month.append(profit_atax[i] == max(profit_atax))
print("Best Month:", best_month)

#9: worst Month 
worst_month = []
for i in range (0, len(profit_atax)):
    worst_month.append(profit_atax[i] == min(profit_atax))
print("Worst Month:", worst_month)

#10: converting all the units into unit of 1000 dollar

revenue_1000 = [round(i/1000, 2) for i in revenue]
expenses_1000 = [round(i/1000, 2) for i in expenses]
profit_1000 = [round(i/1000, 2) for i in profit] 
profit_atax_1000 = [round(i/1000, 2) for i in profit_atax]


revenue_1000 = [int(i) for i in revenue_1000]
expenses_1000 = [int(i) for i in expenses_1000]
profit_1000 = [int(i) for i in profit_1000] 
profit_atax_1000 = [int(i) for i in profit_atax_1000]


print("Revenue:", revenue_1000)
print("Expenses:", expenses_1000)
print("Profit:", profit_1000)
print("Profit_atax:", profit_atax_1000) 


```
